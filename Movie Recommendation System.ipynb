### Data Transforming & Labeling

#### Helper Function

```python
#make a function that extract all names, dirctor name etc from list of dictionary and convert it into a list of strings
def names(obj):
    actor_name = []
    for _name in ast.literal_eval(obj):
        actor_name.append(_name["name"])
    return actor_name

def director(obj):
    director = []
    for _name in ast.literal_eval(obj):
        if _name["job"] == "Director":
            director.append(_name["name"])
            break
    return director

def actors(obj):
    actors = []
    counter = 0
    for _name in ast.literal_eval(obj):
        if counter != 3:
            actors.append(_name["name"])
            counter +=1
        else:
            break
    return actors
```

```python
movies["genres"] = movies["genres"].apply(names)
```

```python
movies["keywords"] = movies["keywords"].apply(names)
```

```python
movies["crew"] = movies["crew"].apply(director)
```

```python
movies["cast"] = movies["cast"].apply(actors)
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

# Apply Machine Learning Model & Traning

### ML Model

```python
from sklearn.feature_extraction.text import TfidfVectorizer as tv
tfidf = tv()
```

```python
tf_idf = tfidf.fit_transform(new_movies['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(tf_idf)
```

```python
similarities.shape
```

```python
def recomd(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    dist = similarities[movie_index]
    movie_list = sorted(list(enumerate(dist)), reverse=True, key=lambda x:x[1])[1:6]
    for mov_ind in movie_list:
        print(new_movies.iloc[mov_ind[0]].title)        
```

```python
recomd("Spectre")
```
